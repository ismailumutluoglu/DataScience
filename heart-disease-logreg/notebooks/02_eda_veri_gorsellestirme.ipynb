```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/heart.csv")
```

### ðŸŽ¯ Hedef DeÄŸiÅŸkenin DaÄŸÄ±lÄ±mÄ±
Bu grafik, veri setinde kaÃ§ kiÅŸinin kalp hastalÄ±ÄŸÄ±na sahip olduÄŸunu (1) ve olmadÄ±ÄŸÄ±nÄ± (0) gÃ¶sterir.


```python
sns.countplot(x='target', data=df)
plt.title("Kalp HastalÄ±ÄŸÄ± DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("Kalp HastalÄ±ÄŸÄ± (0 = Yok, 1 = Var)")
plt.ylabel("KiÅŸi SayÄ±sÄ±")
plt.show()
```

### ðŸ‘´ YaÅŸ DaÄŸÄ±lÄ±mÄ±
Grafikte yaÅŸ arttÄ±kÃ§a kalp hastalÄ±ÄŸÄ± riskinin de arttÄ±ÄŸÄ± gÃ¶zleniyor. Bu, yaÅŸÄ±n model iÃ§in Ã¶nemli bir deÄŸiÅŸken olduÄŸunu gÃ¶sterir.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='target', bins=30, kde=True)
plt.title("YaÅŸ ve Kalp HastalÄ±ÄŸÄ± Ä°liÅŸkisi")
plt.xlabel("YaÅŸ")
plt.ylabel("KiÅŸi SayÄ±sÄ±")
plt.show()
```

## Cinsiyet ve Kalp HastalÄ±ÄŸÄ± arasÄ±ndaki iliÅŸki

```python
sns.countplot(x='sex', hue='target', data=df)
plt.title("Cinsiyet ve Kalp HastalÄ±ÄŸÄ±")
plt.xlabel("Cinsiyet (0 = KadÄ±n, 1 = Erkek)")
plt.ylabel("KiÅŸi SayÄ±sÄ±")
plt.show()
```
